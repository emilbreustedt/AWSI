```python
import matplotlib.pyplot as plt

x = [2020,2050]
menschen = [8, 9.7]
foodwaste = [880, 1498]
fleisch = [3.8, 4.6075]

fig, ax1 = plt.subplots()
fig.suptitle('Trendentwicklungen ohne Insekten bis 2050')

ax1.plot(x, menschen, color='tab:red')
ax1.plot(x, fleisch, color='tab:green')
ax1.plot([], [], 'tab:blue')

ax1.tick_params(axis='y')
ax1.legend(['Milliarden Menschen', 'GHG Fleisch', 'Foodwaste in Mio. Tonnen'])
ax1.set_xticks((2020, 2050))

ax2 = ax1.twinx()
ax2.set_ylabel('GHG Fleisch Mio. Tonnen  CO2e/Jahr', color='tab:blue')  # we already handled the x-label with ax1
ax2.plot(x, foodwaste, color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

fig.tight_layout()

plt.plot()
```
